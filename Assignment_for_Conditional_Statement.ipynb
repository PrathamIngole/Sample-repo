# Q7. Write a Python program to check whether a triangle is Equilateral, Isosceles or Scalene.

Test Data :
50 50 60

Expected Output :
This is an isosceles triangle.

```python
print("Calculating a Triangle is Equilateral, Isosceles or Scalene.")
print("Enter Sides Of Triangle")

a = int(input("Side a = "))
b = int(input("Side b = "))
c = int(input("Side c = "))

if a == b and b == c : 
    print("This is an Equilateral triangle.")
elif (a==b or b == c or a == c) : 
    print("This is an isosceles triangle.")
else :
    print("This is an Scalene triangle.")
```

# Q16. Given three Cartesian coordinates, the task is to check if a right-angled triangle can be formed by the given coordinates. If it is possible to create a right-angled triangle, print Yes. Otherwise, print No.

Examples 1: 

Input: X1=0, Y1=5, X2=19, Y2=5, X3=0, Y3=0 

Output: Yes 


Explanation: 

Length of side connecting points (X1, Y1) and (X2, Y2) is 12.

Length of side connecting points (X2, Y2) and (X3, Y3) is 15. 

Length of side connecting points (X1, Y1) and (X3, Y3) is 9. 

(12)**2 + (9)**2 = 15**2 .

Therefore, a right-angled triangle can be made.

Example 2:

Input: X1=5, Y1=14, X2=6, Y2=13, X3=8, Y3=7 

Output: No

```python
print("Calculating given coordinates are making a triangle or not : ")

x1 = int(input("Enter coordinates of point X1 : "))
y1 = int(input("Enter coordinates of point Y1 : "))

x2 = int(input("Enter coordinates of point X2 : "))
y2 = int(input("Enter coordinates of point Y2 : "))

x3 = int(input("Enter coordinates of point X3 : "))
y3 = int(input("Enter coordinates of point Y3 : "))

d12 = ((x2 - x1)**2 + (y2 - y1)**2)**0.5
d23 = ((x3 - x2)**2 + (y3 - y2)**2)**0.5
d31 = ((x1 - x3)**2 + (y1 - y3)**2)**0.5

# checking is triangle even formed or not 
if (d12 + d23) > d31 and (d23 + d31) > d12 and (d31 + d12) > d23 : 
    if ((d12)**2 + (d23)**2) == ((d31)**2) or  ((d23)**2 + (d31)**2) == ((d12)**2) or ((d31)**2 + (d12)**2) == ((d23)**2): 
        print("Yes, its a Right Angle Triangle.")
    else :
        print("No, its Not a Right Angle Triangle.")
else :
    print("Points does not making triangle (Collinear Points).")
```
