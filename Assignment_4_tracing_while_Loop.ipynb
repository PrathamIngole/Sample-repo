# Q27

```python
x = 15
x ^= x >> 1
x ^= x >> 2
print(x)

# OUTPUT --> 0 

# >> , << shift operators has the higher precedence than ^ Exor operator
# 15 --> 1111 
# x ^= x >> 1 --> 15 ^ (15 >> 1) --> 1111 ^ (1111 >> 1) --> 1111 ^ 0111 = 15 ^ 7 --> 8 
# x ^= x >> 2 --> 8 ^ (8 >> 2) --> 1000 ^ (1000 >> 2) --> 1000 ^ 0010 = 8 ^ 2 --> 10
# Corrected OUTPUT --> 10
```

# Q30.expain what the program is doing?

```python
# Take a and b value

a = int(input("enter a : "))
b = int(input("enter b : "))

while b != 0:
    carry = a & b
    a = a ^ b
    b = carry << 1
print(a)
```
