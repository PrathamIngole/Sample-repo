# Q11. Quadratic Equation Roots
### Write a Python code to calculate the root of a quadratic equation.
Test Data : 1 5 7

Expected Output : 
Root are imaginary;
No solution.

```python
print("Calculation of Roots of a Quadratic Equation.........")
print("Enter the value of a, b and c")

a = int(input("a = "))
b = int(input("b = "))
c = int(input("c = "))

D = (b**2) - (4*a*c)

if D > 0 : 
    print("Distinct Real Roots, two real roots")
elif D == 0 :
    print ("Eactly one Real Root")
else : 
    print ("Roots are imaginary, no solution")
```

# Q14. Triangle Type Determination
### Write a Python code to check whether a triangle is Equilateral, Isosceles or Scalene.
Test Data :
50 50 60

Expected Output :
This is an isosceles triangle.

```python
print("Calculating a Triangle is Equilateral, Isosceles or Scalene.")
print("Enter Sides Of Triangle")

a = int(input("Side a = "))
b = int(input("Side b = "))
c = int(input("Side c = "))

if a == b and b == c : 
    print("This is an Equilateral triangle.")
elif (a==b or b == c or a == c) : 
    print("This is an isosceles triangle.")
else :
    print("This is an Scalene triangle.")
```
