# Q5. find the output of following snippet

```python
x = 0
for i in range(1, 6):
    if i % 2 == 0:
        x += i
    else:
        x -= i
    print(x)

# OUTPUT --> -1 1 -2 2 -3
# Correct answer 
```

# Q6. find the output of following snippet

```python
s = 0
for i in range(1, 11):
    if s % 2 == 0:
        s += i
        print(i, "if  = ", s)
    else:
        s -= i
        print(i, "else = ", s)
print("last s = ",s)

# OUTPUT --> 1 -1 -4  0  5 -1 -8  0  9 -1
```

# Q7. find the output of following snippet

```python
j = 1
for i in range(5, 0, -1):
    if j % 2 != 0:
        j += i
    else:
        break
print(i + j)

# for loop tracing j --> 6 
# OUTPUT --> 10
```

# Q8. find the output of following snippet

```python
a = 1
b = 2

for i in range(1, 6) :
    if i % 2 == 0 :
        a *= b
    else :
        b += a
print(a + b)

# for loop tracing calculation of :
# a --> 3 18 
# b --> 3 6 24
# OUTPUT --> 42
```

# Q9. find the output of following snippet

```python
result = 0

for i in range(1, 6):
    for j in range(1, i+1):
        result += j
        
print(result)

# for loop tracing calculation of result --> 1, 2 4, 5 7 10, 11 13 16 20, 21 23 26 30 35
# OUTPUT --> 35
```
